# Text Preprocessing
Demonstrates text processing using the Porter Stemmer from the nltk library to preprocess a series of textual reviews. First, a PorterStemmer object is instantiated, which is used to "stem" words in the text—essentially reducing words to their root form. The code then applies two lambda functions to a pandas DataFrame column (rev). 

The first lambda function tokenizes each review text into individual words, converts them to lowercase, and applies the PorterStemmer to stem each word. The second lambda function joins the processed words back into a single string, forming a new representation of the reviews with stemmed words. The last line outputs the processed data in a DataFrame format.

```python
ps = PorterStemmer()
rev = rev.apply(lambda x:[ps.stem(i.lower()) for i in word_tokenize(x)]).apply(lambda y: " ".join(y))
rev
```

# Output
The output demonstrates the transformed text, where words are stemmed, and sentences are reconstructed. This technique is common in text mining, search, and other natural language processing tasks to ensure consistency and improve the accuracy of operations like searching or sentiment analysis.

## Text Preprocessing by removing Stopwords from text data
Used the NLTK library to remove stopwords from text data stored in a pandas Series named "rev". It imports the "stopwords" module, retrieves the list of English stopwords, and applies a lambda function to tokenize each string, filter out stopwords (case-insensitive), and rejoin the remaining tokens into a new string. By removing stopwords, it helps to reduce the noise in the text data and focus on the more meaningful words. This preprocessing step is often beneficial for various natural language processing tasks, such as text classification, sentiment analysis, topic modeling, and information retrieval, as it can improve the performance and accuracy of these tasks by removing irrelevant words.


```python
from nltk.corpus import stopwords
st = stopwords.words('english')
rev = rev.apply(lambda x:[i for i in word_tokenize(x) if i.lower() not in st]).apply(lambda y: " ".join(y))
rev
```

# Vectorization
We used the "TfidfVectorizer" class from sklearn.feature_extraction.text to convert a collection of raw text documents (rev) into a matrix of TF-IDF (Term Frequency-Inverse Document Frequency) features. This process is commonly used in text analysis and natural language processing to represent the relative importance of words in a set of documents while considering both the frequency of words and their rarity across the dataset.

The "TfidfVectorizer" object tf1 is instantiated and then fitted to the input data rev with tf1.fit_transform(rev).
